```python
import torch 
from torch import nn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
w=0.7
b=0.3
x=torch.arange(0,1,0.02)
```

```python
y=w*x+b 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_tarin , y_test =train_test_split(x,y,train_size=0.8)
```

```python
def plot_prediction( x_train , y_tarin , x_test , y_test , predection=None):
    plt.scatter(x_train, y_tarin, color='blue', label='Training Data')
    plt.scatter(x_test, y_test, color='red', label='Training Data')
#    if(predection!=None):
    plt.scatter(x_test, predection, color='green', label='Training set ')
```

```python
plot_prediction( x_train , y_tarin , x_test , y_test , predection=None)
```

```python
class LinearReg(nn.Module):
    def __init__(self):
        super(LinearReg,self).__init__()
        self.weights=nn.Parameter(torch.rand(1,dtype=torch.float64) , requires_grad=True)
        self.bias=nn.Parameter(torch.rand(1,dtype=torch.float64),requires_grad=True)
    def forward(self , x)-> torch.tensor:
        return self.weights*x + self.bias
```

```python
torch.manual_seed(42)
model=LinearReg()
```

```python
model = LinearReg()
with torch.inference_mode():
    output = model(x_test)
print(output)
```

```python
list(model.parameters())
```

```python
plt.scatter(x_test,output,c="green")
```

```python
plot_prediction( x_train , y_tarin , x_test , y_test , output)
```
